```python
import cv2 
from IPython.display import Image 
import numpy as np
from matplotlib import pyplot as plt 

```

```python
#Cargar Imagenes
ima1 = cv2.imread("imagenes/m1.jpg",0)
ima2 = cv2.imread("imagenes/m1.jpg",1)
#Crear un nuevo plot
plt.figure()
#Plotear imagenes
plt.subplot(1,2,1);plt.imshow(ima1),plt.title('1')
plt.subplot(1,2,2);plt.imshow(ima2),plt.title('2')

ima21 = cv2.cvtColor(ima1,cv2.COLOR_BGR2RGB)
ima22 = cv2.cvtColor(ima2,cv2.COLOR_BGR2RGB)

plt.figure()
plt.subplot(1,2,1);plt.imshow(ima21),plt.title("3")
plt.subplot(1,2,2);plt.imshow(ima22),plt.title("4")
```

```python
#Podemos observar el tamaño de los distintos filtros aplicado a
#las distintas imagenes
print("alpha",ima1.shape)
print("BGR",ima2.shape)
print("Negro",ima21.shape)
print("Color",ima22.shape)
```

```python
#Con hsv podemos ver las distintas tonalidades de la imagen.
img_hsv = cv2.cvtColor(ima22,cv2.COLOR_RGB2HSV)
umbral_bajo = (70,100,100)
umbral_alto = (80,255,255)
#Una mascara es una región de la imagen que queremos separar
#Observaremos por tanto solo la parte donde hay verde en la imagen.
mask = cv2.inRange(img_hsv,umbral_bajo,umbral_alto)
res = cv2.bitwise_and(ima22,ima22,mask=mask)
plt.figure()
#cmap nos permite cambiar el color de la binarizaciónn, lo cambiamos a gray
plt.subplot(1,2,1);plt.imshow(mask,cmap="gray"),plt.title("3")
plt.subplot(1,2,2);plt.imshow(res),plt.title("4")

```

```python
#Separa el Color Azul
img_hsv = cv2.cvtColor(ima22,cv2.COLOR_RGB2HSV)
umbral_bajo = (110,100,100)
umbral_alto = (130,255,255)
#Una mascara es una región de la imagen que queremos separar
#Observaremos por tanto solo la parte donde hay verde en la imagen.
mask = cv2.inRange(img_hsv,umbral_bajo,umbral_alto)
res = cv2.bitwise_and(ima22,ima22,mask=mask)
plt.figure()
#cmap nos permite cambiar el color de la binarizaciónn, lo cambiamos a gray
plt.subplot(1,2,1);plt.imshow(mask,cmap="gray"),plt.title("3")
plt.subplot(1,2,2);plt.imshow(res),plt.title("4")
```
